```python
#THIS CELL FLATTENS (rollout) a table to a column
#eg.
#input = 3 x 4 matrix 
        # A B C 
        # E F G
        # H I J
#output = [A B C D E F G H I J] (in a column)

#In matlab, this is essentially A -> A(:) operation
    # in matlab, the data is rolled out into a column
    # in numpy, the data is flatten into a row (ie need to transpose to get column)

import pandas as pd
import numpy as np

#load file as a DataFrame first; easy way to load data
df = pd.read_excel("flatten_input.xlsx", header = None) 
    #This is the file with the table
    #"header = None" such that df *includes* values from 1st row. 
    #by default, the 1st row is assumed to be headers (ie "titles")
pd.set_option('display.max_columns', None) 
    #this is useful if you need to preview data; might not be needed otherwise

#convert df to an array, in order to carry out the "rolling" 
df_array = df.to_numpy()

#flatten the array
flatten_array = df_array.flatten()

#tranpose resulting single row to a single column
flatten_array_transpose = np.transpose(flatten_array)

#convert data from array to DataFrame; easy way to output into excel/readable/exportable format
df_flatten_array = pd.DataFrame(flatten_array_transpose)
df_flatten_array.to_excel('flatten_output.xlsx', index = False, header = False)
    # "index = false" as do not want index tagged to the rows of the output "matrix"

df_flatten_array.head(5)
print(len(df_flatten_array))
```
