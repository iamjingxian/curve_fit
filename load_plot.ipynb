```python
#THIS CELL IS TO CHOOSE DATA TO PLOT

import numpy as np

#initialise df for diff timepoints
#need to create a DF for each time series
df_30min = df[['conc (ng/mL)','470_30min','526_30min']]
df_1h = df[['conc (ng/mL)','470_1h','526_1h']]
df_2h = df[['conc (ng/mL)','470_2h','526_2h']]
df_4h = df[['conc (ng/mL)','470_4h','526_4h']]


#Derive the ratio 526/470
#note: on the instrument, "auto cutoff" function is set to ON
df_30min['526/470']=df['526_30min']/df['470_30min']
df_1h['526/470']=df['526_1h']/df['470_1h']
df_2h['526/470']=df['526_2h']/df['470_2h']
df_4h['526/470']=df['526_4h']/df['470_4h']

#choose only columns with conc and 526/470 to display
df_30min=df_30min[['conc (ng/mL)', '526/470']]
df_1h=df_1h[['conc (ng/mL)', '526/470']]
df_2h=df_2h[['conc (ng/mL)', '526/470']]
df_4h=df_4h[['conc (ng/mL)', '526/470']]

#THIS SECTION PLOTS THE DATA

from IPython.display import display_html

#display the tables of time series side by side
#need to include one "df styler" per time series
#"inline' means dispaly within this notebook page
df1_styler = df_30min.style.set_table_attributes("style='display:inline'").set_caption('30min')
df2_styler = df_1h.style.set_table_attributes("style='display:inline'").set_caption('1h')
df3_styler = df_2h.style.set_table_attributes("style='display:inline'").set_caption('2h')
df4_styler = df_2h.style.set_table_attributes("style='display:inline'").set_caption('4h')

#This is the code that displays the tables; 
#the codes above set the parameters of the tables to display
display_html(df1_styler._repr_html_()
             +df2_styler._repr_html_()
             +df3_styler._repr_html_()
             +df4_styler._repr_html_(), raw=True)
```
